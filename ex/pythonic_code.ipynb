```python
case_1 = ["A","B","C"]
case_2 = ["D","E","A"]
result_1 = [i+j for i in case_1 for j in case_2]
print("result_1: ", result_1)
result_2 = [i+j for i in case_1 for j in case_2 if not (i==j)]
print("result_2 : ", result_2)
```
