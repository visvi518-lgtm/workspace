```python
class Animal:
    def __init__(self, name):
        self.name = name

    def talk(self):
        raise NotImplementedError("Subclass must implement abstract method")
```

```python
class Cat(Animal):
    def talk(self):
        return "Meow!"

class Dog(Animal):
    def talk(self):
        return "woof! woof!"
```

```python
animals =[Cat("Missy"),
          Cat("Mr.Mistoffeless"),
          Dog("Lassie")]
```

```python
for animal in animals:
    print(animal.name+': ' +animal.talk())
```
