```python
s1 = set([1,2,3,4,5])
s2 = set([3,4,5,6,7])
print(type(s1))
print(type(s2))
print("union : ", s1.union(s2))
print("intersection : ", s1.intersection(s2))
print("difference : ", s1.difference(s2))
print(s1-s2)
```

```python
from collections import deque
deque_list = deque()
for i in range(5):
    deque_list.append(i)
deque_list.appendleft(10)
deque_list
```

```python
deque_list.rotate(1)
deque_list
```
