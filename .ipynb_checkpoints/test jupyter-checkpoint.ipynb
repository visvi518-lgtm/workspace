```python
for i in range(10):
    if i == 5:
        break
    print(i)
print("EOP")
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
print("EOP")
```

```python
for i in range(10):
    print(i,)
else:
    print("EOP")
```

```python
sentence = " i love you"
reverse_sen = ""
for char in sentence:
    print("REVERSE #1", reverse_sen)
    reverse_sen = char +reverse_sen
    print("REVERSE #2", reverse_sen)
```

```python
decimal = int(input("십진수를 입력해주세요 : "))
result = ""

while decimal >0:
    remainder = decimal % 2
    decimal = decimal //2
    result = str(remainder) + result
print(result)
```
