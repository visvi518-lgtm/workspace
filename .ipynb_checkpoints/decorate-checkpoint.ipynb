```python
def star(func):
    def inner(*args, **kwargs):
        print("*"*30)
        func(*args, **kwargs)
        print("*"*30)
    return inner

@star
def printer(msg):
    print(msg)
printer("hello")
```

```python
def star(func):
    def inner(*args, **kwargs):
        print(args[1]*30)
        func(*args, **kwargs)
        print(args[1]*30)
    return inner

@star
def printer(msg, mark):
    print(msg)
printer("hello", "T")
```

```python
def generate_power(exponent):
    def wrapper(f):
        def inner(*args):
            result = f(*args)
            return exponent**result
        return inner
    return wrapper

@generate_power(2)
def raise_two(n):
    return n**2
```

```python
print(raise_two(7))
```
